### Desafio 8

Escreva um script Python para encontrar as 10 palavras mais longas em um arquivo de texto.
O arquivo .txt está localizado na mesma pasta do projeto (**texto.txt**).

```python
# Seu código
```

```python
# Importando bibliotecas
import pandas as pd
import requests
import re
```

```python
# Importando texto do GitHub
url = 'https://raw.githubusercontent.com/leticiaplang/br-python-challenges/master/texto.txt'
req = requests.get(url)
text = req.text
print(text)
```

### Modelo 1 - NLTK

```python
text2 = text
text2
```

```python
# importações de bibliotecas
import nltk
nltk.download()
```

```python
from nltk.tokenize import word_tokenize
from string import punctuation
```

```python
# Captando e transformando as palavras em minúsculas
words_nltk = word_tokenize(text2.lower())
words_nltk[:11]
```

```python
# looping for para gerar lista com os tamanhos de cada palavra
size_nltk = []
for n in words_nltk:
  size_nltk.append(len(n))
```

```python
# Gerandp um dataframe com as duas listas 
df_nltk = pd.DataFrame({'words':words_nltk, 'size':size_nltk})
df_nltk
```

```python
# Realizando a ordenação das 10 maiores palavras
df_nltk.nlargest(10, 'size')
```

#### Modelo 2 - Pandas

```python
# Poderíamos utilizar Regex também
text3 = text.replace('?','').replace('\r\n','').replace('.',' ').replace(',', '').replace(')', '').replace('(', '')
text3
```

```python
# Gerando uma lista de palavras
words = [n for n in text3.split(' ')]
words[:11]
```

```python
# Gerando uma lista com os tamanhos de cada palavra
size = [len(n) for n in words]
size[:11]
```

```python
# Gerando um dataframe com as duas listas
df = pd.DataFrame({'words':words, 'size':size})
df
```

```python
# Opção 1 - ordenando as 10 maiores palavras
df.nlargest(10, 'size')
```

```python
# Opção 2 - ordenando as 10 maiores palavras
df.sort_values(by='size', ascending=False).head(10)
```
